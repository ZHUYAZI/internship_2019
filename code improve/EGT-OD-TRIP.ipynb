```
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.geometry import Point
```

```
file_deplacement ="/content/drive/My Drive/Colab Notebooks/stage2019_improvement/data_egt/Deplacements_semaine.csv"
file_shape ="/content/drive/My Drive/Colab Notebooks/stage2019_improvement/data_egt/Carreaux_shape_mifmid/carr100m.shp"
file_station_coordinate ="/content/drive/My Drive/Colab Notebooks/stage2019_improvement/data_egt/gare_ferre_corrodonne.csv"
```

```
deplacement=pd.read_csv(file_deplacement,delimiter=',',encoding='iso 8859-1')
deplacement
```

```
carre_shape = gpd.read_file(file_shape,encoding='latin-1')
carre_shape
```

```
h = carre_shape[["IDENT","X100","Y100"]].copy()
h
```

```
geometry = [Point(xy) for xy in zip(h.X100, h.Y100)]
geometry
```

```
hg = gpd.GeoDataFrame(h, geometry=geometry)
hg
```

```
hg.crs = {'init' :'epsg:27561'}  
# ^ comment out to get a "Cannot transform naive geometries" error below

# project to merkator
hg = hg.to_crs({'init': 'epsg:4326'})
hg
```

```
hg["lon"]=hg.geometry.x
hg["lat"]=hg.geometry.y
hg
```

```
dict_geometry={}
for index,row in hg.iterrows():
    print(index)
    dict_geometry[row['IDENT']]=(row['lon'],row['lat'])
dict_geometry
```

```
station_location = pd.read_csv(file_station_coordinate,delimiter=';',encoding='latin-1')
station_location
```

```
Original = deplacement.groupby(["ORC"]).size().reset_index(name='counts')
Original[["lon","lat"]] = pd.DataFrame([[np.nan,np.nan]])
Original
```

```
for index,row in Original.iterrows():
  print(index)
  Original.loc[index,"lon"],Original.loc[index,"lat"] = dict_geometry[row["ORC"]][0],dict_geometry[row["ORC"]][1]
```

```
Original
```

```
station_location["lattitude"] = station_location["Geo Point"].apply(lambda x : float(x.split(",")[0]))
station_location["longtitude"] = station_location["Geo Point"].apply(lambda x : float(x.split(",")[1]))
station_location
```

```
# mpu.haversine_distance((lat1, lon1), (lat2, lon2))

for i,row_i in Original.iterrows():
  lat1 = Original.iloc[i]["lat"]
  lon1 = Original.iloc[i]["lon"]
  pos1=(lat1,lon1)
  dist=float("inf")
  station = " "
  station_lat = 0
  station_lon = 0
  for index,row in station_location.iterrows():
    if dist > mpu.haversine_distance(pos1, (row["lattitude"], row["longtitude"])):
      dist = mpu.haversine_distance(pos1, (row["lattitude"], row["longtitude"]))
      station = row["nomlong"]
      station_lat = row["lattitude"]
      station_lon = row["longtitude"]

  Original.loc[i,"dist"] = dist
  Original.loc[i,"nomlong"] = station
  Original.loc[i,"station_lat"] = station_lat
  Original.loc[i,"station_lon"] = station_lon
  print(i/len(Original))
```
