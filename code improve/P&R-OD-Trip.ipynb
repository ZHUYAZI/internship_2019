```
import pandas as pd
import numpy as np
import mpu
```

```
parc_relais_file = "/content/drive/My Drive/Colab Notebooks/stage2019_improvement/parcs-relais-idf.csv"
O_file = "/content/drive/My Drive/Colab Notebooks/stage2019_improvement/Original.csv"
D_file ="/content/drive/My Drive/Colab Notebooks/stage2019_improvement/Destination.csv"
```

```
parking =pd.read_csv(parc_relais_file,delimiter=';',encoding='iso 8859-1')
parking
```

```
parking["lattitude"] = parking["Geo Point"].apply(lambda x : float(x.split(",")[0]))
parking["longtitude"] = parking["Geo Point"].apply(lambda x : float(x.split(",")[1]))
parking
```

```
O_data = pd.read_csv(O_file,delimiter=';',encoding='iso 8859-1')
O_data=O_data.groupby(["nomlong","station_lat","station_lon"],as_index=False)["counts"].sum()
O_data
```

```
for index,row in O_data.iterrows():
  print(index/len(O_data))
  lat1 = row["station_lat"]
  lon1 = row["station_lon"]
  dist = float("inf")
  name_parking_possible =" "
  lat_park= 0
  lon_park =0
  for i , j in parking.iterrows():
    lat2 = j["lattitude"]
    lon2 = j["longtitude"]
    if dist > mpu.haversine_distance((lat1,lon1), (lat2,lon2)):
      dist =mpu.haversine_distance((lat1,lon1), (lat2,lon2))
      name_parking_possible =j["NOM_PR"]
      lat_park = lat2
      lon_park =lon2

  O_data.loc[index,"dist_to_PR"] = dist
  O_data.loc[index,"PR_NAME"] = name_parking_possible
  O_data.loc[index,"lat_park"] = lat_park
  O_data.loc[index,"lon_park"] = lon_park
```

```
O_data
```

```
O_park = O_data[O_data["dist_to_PR"]<=1].copy().reset_index(drop=True)
```

```
O_park
```

```
O_park_count=O_park.groupby(["PR_NAME","lat_park","lon_park"],as_index=False)["counts"].sum()
O_park_count
```

```
O_park_count.to_csv("O-P&R.csv",sep=";",index=False)
```
