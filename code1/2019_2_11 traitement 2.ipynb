```python
import datetime
def change_time(time):
    return datetime.datetime.strptime(time,'%Y-%m-%d %H:%M:%S')

def selet_time(start,end,data):
    start_time=change_time(start)
    end_time=change_time(end)
    new_data=data[(data['HEUR_VALD']>start_time)&(data['HEUR_VALD']<end_time)]
    return new_data
```

```python
data_RERA_Entree=data_RERA[data_RERA['LIBL_NATR_VALD'].str.contains('Entree')]
data_RERA_Sortie=data_RERA[data_RERA['LIBL_NATR_VALD'].str.contains('Sortie')]
```

```python
def OD_matrix(start,end,stop_choice):
    new_data=selet_time(start,end,data_RERA_Entree)
    new_data=new_data[new_data['LIBL_LONG_ARRT_STTN']==stop_choice]
    print("the number of the people enter {gare} during {start} and {end} is : {number}".format(gare=stop_choice,start=start,end=end,number=len(new_data)))
    final_data=selet_time(start,end,data_RERA_Sortie)
    final_data=final_data[final_data['NUMR_CART'].isin(new_data['NUMR_CART'])]
    print("the number of the people exit during {start} and {end} is : {number}".format(start=start,end=end,number=len(final_data)))  
    counts=count(final_data['LIBL_LONG_ARRT_STTN'])
    return pd.DataFrame([i for i in counts.values()],index=counts.keys(),columns=[stop_choice])
```
