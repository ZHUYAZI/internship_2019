```python
import time
from functools import wraps
def decorate(func):
    @wraps(func)
    def calcul_time_func(*args,**kwargs):
        start=time.time()
        result=func(*args,**kwargs)
        end=time.time()
        print(func.__name__,' needs time ',end-start,' s')
        return result
    return calcul_time_func
```

```python
@decorate
def get_trip_max(trip):
    trip_max={}
    for index,value in trip[['client_id','trip_id']].iterrows():
        trip_max[value[0]]=value[1]
    
    return trip_max  
```

```python
trip_max=get_trip_max(trip)
```

```python
import datetime
```

```python
trip['Ot']=trip[trip['Ot'].notnull()]['Ot'].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
trip['Dt']=trip[trip['Dt'].notnull()]['Dt'].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
```

```python
trip
```

```python
time_valide=trip[~trip['Ot'].isnull()].copy()
time_valide=time_valide[~time_valide['Dt'].isnull()]
time_valide
```

```python
time_valide['transfer_time']=(time_valide['Dt']-time_valide['Ot']).dt.total_seconds()
```

```python
time_valide
```

```python
plt.figure(figsize=(15,5))
plt.hist(time_valide[time_valide['transfer_time']<=60000]['transfer_time'], bins=400)
plt.title('distribution du temps de transfer')
plt.xlabel('transfer time')
plt.ylabel('Count')
plt.show()
```
