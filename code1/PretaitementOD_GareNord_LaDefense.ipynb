```python
def output2(data,T):   # T means the period between each O and D
    transfer_id = 0      #count the trandfer like index
    trip_id=1           #count the trip
    trips = pd.DataFrame([], columns=columns_n)  #create dateframe for conserve the result
    count_number=0      #count the number of the validation of each numr_cart
    if_merge=False
    merge_time=0
    for i in range(len(data)-1):
        first = data.loc[i]
        second = data.loc[i+1]       #compare each 2 line in the dataset
        if first['NUMR_CART']==second['NUMR_CART']:
            if 'SORTIE' in first['LIBL_NATR_VALD'] and not if_merge :
                trips.loc[transfer_id,['trip_id', 'client_id','date', 'D','Dt','mode','nature_D']] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],first['LIBL_LONG_ARRT_STTN'],first['HEUR_VALD'],first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD']]
                if 'SORTIE' in second['LIBL_NATR_VALD']:#renew the following ligne
                    trips.loc[transfer_id+1,['O', 'Ot','nature_O']]=[first['LIBL_LONG_ARRT_STTN'],first['HEUR_VALD'],first['LIBL_NATR_VALD']]
                #else out->in do nothing
            if_merge=False
            if 'ENTREE' in first['LIBL_NATR_VALD']:
                trips.loc[transfer_id] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],
                                      first['LIBL_LONG_ARRT_STTN'], second['LIBL_LONG_ARRT_STTN'],
                                      first['HEUR_VALD'],second['HEUR_VALD'],first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD'],
                                      second['LIBL_NATR_VALD']]
                if 'SORTIE' in second['LIBL_NATR_VALD']:
                    if_merge=True
                    merge_time+=1
                    trips.loc[transfer_id+1,['O', 'Ot','nature_O']]=[second['LIBL_LONG_ARRT_STTN'],second['HEUR_VALD'],second['LIBL_NATR_VALD']]
                    
                    
            if not if_merge:
                transfer_id +=1
                count_number+=1
            
            if (second['HEUR_VALD']-first['HEUR_VALD']).total_seconds()>=T and not if_merge:
                    trip_id+=1
        else:
            if count_number==0 and not if_merge : #delete the data who has the unique inforamtion of validation 
                continue
            first = data.loc[i]           #define the D of the last transfer
            second = data.loc[i-count_number-merge_time]
            if if_merge :
                trips=trips.drop(index=transfer_id+1)
                if 'SORTIE' in second['LIBL_NATR_VALD']:
                    trips.loc[transfer_id-count_number,['O','nature_O']]=[first['LIBL_LONG_ARRT_STTN'],first['LIBL_NATR_VALD']]
                
            else:
                if 'SORTIE' in first['LIBL_NATR_VALD']:
                    trips.loc[transfer_id,['trip_id', 'client_id','date', 'D','Dt','mode','nature_D']] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],first['LIBL_LONG_ARRT_STTN'],first['HEUR_VALD'],first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD']]
                    if 'SORTIE' in second['LIBL_NATR_VALD']:
                        trips.loc[transfer_id-count_number,['O','nature_O']]=[first['LIBL_LONG_ARRT_STTN'],first['LIBL_NATR_VALD']]
               
                else:#'entree' in first
                    if 'ENTREE' in second['LIBL_NATR_VALD']:
                        trips.loc[transfer_id] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],
                                      first['LIBL_LONG_ARRT_STTN'], second['LIBL_LONG_ARRT_STTN'],
                                      first['HEUR_VALD'],np.nan,first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD'],
                                      second['LIBL_NATR_VALD']]
                    else:
                        trips.loc[transfer_id,['trip_id', 'client_id','date', 'O','Ot','mode','nature_O']] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],first['LIBL_LONG_ARRT_STTN'],first['HEUR_VALD'],first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD']]
            
                    
            if_merge=False 
            transfer_id +=1
            count_number=0
            merge_time=0
            trip_id=1
    # handle with the last line in dataset
    
    if if_merge:
        trips.drop(index=[transfer_id+1],inplace=True)
    if count_number!=0 and not if_merge:
       
        first = data.loc[i+1]
        second = data.loc[i+1-count_number-merge_time]
                                                  
        if 'SORTIE' in first['LIBL_NATR_VALD']:
            trips.loc[transfer_id,['trip_id', 'client_id','date', 'D','Dt','mode','nature_D']] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],first['LIBL_LONG_ARRT_STTN'],first['HEUR_VALD'],first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD']]
            if 'SORTIE' in second['LIBL_NATR_VALD']:
                trips.loc[transfer_id-count_number,['O','nature_O']]=[first['LIBL_LONG_ARRT_STTN'],first['LIBL_NATR_VALD']]

        else:#'entree' in first
            if 'ENTREE' in second['LIBL_NATR_VALD']:
                trips.loc[transfer_id] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],
                                  first['LIBL_LONG_ARRT_STTN'], second['LIBL_LONG_ARRT_STTN'],
                                  first['HEUR_VALD'],np.nan,first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD'],
                                  second['LIBL_NATR_VALD']]
            else:
                trips.loc[transfer_id,['trip_id', 'client_id','date', 'O','Ot','mode','nature_O']] = [trip_id,first['NUMR_CART'],first['JOUR_EXPL_VALD'],first['LIBL_LONG_ARRT_STTN'],first['HEUR_VALD'],first['LIBL_LONG_MOD'],first['LIBL_NATR_VALD']]

    return trips 
```
