```python
from osmnx.utils import great_circle_vec
```

```python
lng1=data1.iloc[0]['geometry'].x
lat1=data1.iloc[0]['geometry'].y
lng2=data1.iloc[1]['geometry'].x
lat2=data1.iloc[1]['geometry'].y
```

```python
great_circle_vec(lat1, lng1, lat2, lng2, earth_radius=6371009)
```

```python
packing=[]
station=[]
limit=1000    #distance between gare et parc should choose a value
for i,rowi in data2.iterrows():  #parcing
    for j,rowj in data1.iterrows():  #gare rer A
        lng1=data2.iloc[i]['geometry'][0].x
        lat1=data2.iloc[i]['geometry'][0].y
        lng2=data1.iloc[j]['geometry'].x
        lat2=data1.iloc[j]['geometry'].y
        distance=great_circle_vec(lat1, lng1, lat2, lng2, earth_radius=6371009)
        if distance <=limit:
            packing.append(i)
            station.append(j)
            break
```

```python
date_save_parc=data2.iloc[packing]
date_save_parc.reset_index(drop=True)
```

```python
import matplotlib.patches as mpatches
f, ax = plt.subplots(1,figsize=(15,20))
line_up=data1.plot(axes=ax)
line_down=date_save_parc.plot(axes=ax,color='r',marker='D')
red_patch = mpatches.Patch(color='red', label='parc relais')
blue_patch = mpatches.Patch(color='blue', label='station de RER A')
plt.legend(handles=[red_patch, blue_patch])
plt.title('Parc relais et station de RER A limite 1 km')
```
