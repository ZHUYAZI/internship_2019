```python
treat_data=no_data.copy()
```

```python
treat_data['Ot']=treat_data['date']+str(' ')+treat_data['Ot']
```

```python
treat_data['Dt']=treat_data['date']+str(' ')+treat_data['Dt']
```

```python
from datetime import datetime
```

```python
treat_data['Ot']=treat_data['Ot'].apply(lambda x:datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
treat_data['Dt']=treat_data['Dt'].apply(lambda x:datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
```

```python
treat_data.rename(columns={'trip_id':'transfer_number'}, 
                 inplace=True)
```

```python
treat_data
```

```python
for index,row in treat_data.iterrows():
    treat_data.loc[index,'transfer_number']=dict_id2[row['client_id']]
     
```

```python
result=treat_data.sort_values(['transfer_number','client_id','Ot'])
result=result.reset_index(drop=True)
result
```

```python
(result.iloc[0]['Dt']-result.iloc[0]['Ot']).total_seconds()
```

```python
trip_time=[]
for key,value in dict_id2.items():
    n1=result[result['client_id']==key].index.values.astype(int)[0]
    n2=n1+value-1
    time=(result.loc[n2,'Dt']-result.loc[n1,'Ot']).total_seconds()
    if time <0:
        time+=24*3600
    trip_time.append(time)
```
