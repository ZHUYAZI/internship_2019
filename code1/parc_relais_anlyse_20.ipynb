```python
import math
def distance_from_table(ident1,ident2):
    x1=float(test[test['IDENT']==ident1].X100)
    y1=float(test[test['IDENT']==ident1].Y100)
    x2=float(test[test['IDENT']==ident2].X100)
    y2=float(test[test['IDENT']==ident2].Y100)
    return get_dist(x1,y1,x2,y2)

def get_dist(x1,y1,x2,y2):
    return math.sqrt((x1-x2)**2+(y1-y2)**2)
```

# 3. SVM supervise learning for data1_true,data2_true,data3_false

```python
import pandas as pd
import numpy as np
```

```python
data1_true=pd.read_csv('C:\\Users\\76571\\Desktop\\stage\\data\\train\\data1_true.csv',delimiter=';',encoding='iso 8859-1')
```

```python
data2_true=pd.read_csv('C:\\Users\\76571\\Desktop\\stage\\data\\train\\data2_true.csv',delimiter=';',encoding='iso 8859-1')
```

```python
data3_false=pd.read_csv('C:\\Users\\76571\\Desktop\\stage\\data\\train\\data3_false.csv',delimiter=';',encoding='iso 8859-1')
```

```python
(len(data1_true)+len(data2_true))/(len(data3_false)+len(data1_true)+len(data2_true))
```

```python
Data_Train=pd.concat([data1_true,data2_true,data3_false],ignore_index=True)
Data_Train
```

```python
X = Data_Train.drop('parking', axis=1)
y = Data_Train['parking']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
y=np.array(y.astype(float))
y
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(random_state=42)
```

```python
X_res,y_res=sm.fit_resample(X,y)
```

```python
X_res.shape
```

```python
y_res.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size = 0.20)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='rbf',class_weight='balanced')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
