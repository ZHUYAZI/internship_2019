```python
time=['2019-02-11 03:00:00',
      '2019-02-12 03:00:00',
      '2019-02-13 03:00:00',
      '2019-02-14 03:00:00',
      '2019-02-15 03:00:00',
      '2019-02-16 03:00:00',
      '2019-02-17 03:00:00',
      '2019-02-18 03:00:00',
     ]
import datetime

time_limit=list(map(lambda x:datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'),time))
```

```python
time_limit
```

```python
#chose the 2019/02/11 3:00 to  2019/02/12 3:00 in the database 
%%time
chunks=pd.read_csv(r'C:\Users\yazi.zhu\Documents\enpc_2019.txt',delimiter=";",encoding='iso 8859-1',chunksize=10_000)
db=sqlite3.connect('2019_02_11')
for chunk in chunks:
    
    if '2019-02-11' in list(chunk['JOUR_EXPL_VALD']) or '2019-02-12' in list(chunk['JOUR_EXPL_VALD']):
        chunk['HEUR_VALD']=chunk['JOUR_EXPL_VALD']+str(' ')+chunk['HEUR_VALD'].apply(lambda x:x.split(' ')[0])
        chunk['HEUR_VALD']=pd.to_datetime(chunk['HEUR_VALD'],format='%Y-%m-%d %H:%M:%S')
        chunk=chunk[chunk['HEUR_VALD']<time_limit[1]]
        chunk=chunk[['NUMR_CART',
                     'HEUR_VALD',
                     'JOUR_EXPL_VALD',
                     'LIBL_LONG_ARRT_STTN',
                     'LIBL_LONG_MOD',
                     'LIBL_NATR_VALD',
                     'COOR_X_PHSQ',
                     'COOR_Y_PHSQ']]
        chunk.to_sql('2019_02_11',db,if_exists='append')
    
   
        
         
```
