# preprocessing data 

```python
dict_geometry={}
for index,row in test.iterrows():
    dict_geometry[row['IDENT']]=(row['X100'],row['Y100'])
dict_geometry
```

```python
import math
def get_dist(x1,y1,x2,y2):
    return float(math.sqrt((x1-x2)**2+(y1-y2)**2))
```

```python
import time
def calcul_dist(data,label1,label2):
    dist=[]
    start=time.time()
    for index,row in data.iterrows():
        try:
            x1=dict_geometry[row[label1]][0]
            y1=dict_geometry[row[label1]][1]
            x2=dict_geometry[row[label2]][0]
            y2=dict_geometry[row[label2]][1]
            dist.append(get_dist(x1,y1,x2,y2))
            end=time.time()
            print('index=',index,' time= ',end-start,' sec ')
        except KeyError:
            dist.append(0)
            end=time.time()
            print('index=',index,' time= ',end-start,' sec ')

    return dist
```

```python
data_negatif_1['Ot']=data_negatif_1['ORH']*3600+data_negatif_1['ORM']*60
data_negatif_1['Dt']=data_negatif_1['DESTH']*3600+data_negatif_1['DESTM']*60
data_negatif_1['DUREE']=data_negatif_1['DUREE']*60
data_negatif_1
```

```python
dist=calcul_dist(data_negatif_1,'ORC','DESTC')
data_negatif_1['dist']=dist
```
